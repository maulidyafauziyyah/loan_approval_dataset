```python
# import library
import pandas as pd
import joblib
import json
import numpy as np
```

```python
# load model
model = joblib.load('model.pkl')
with open('feature.json','r') as f:
    feature = json.load(f)
```

```python
# Membuat variable dummy (penggunaan radint: untuk membuat data random)
data_inf = {
    'loan_id'  : np.random.randint(0,100000,1000), 
    'no_of_dependents'      : np.random.randint(0,100000,1000) , 
    'education': ['Graduate' for _ in range(1000)] , 
    'self_employed'  : ['No' for _ in range(1000)] , 
    'income_annum': np.random.randint(0,100000,1000),
    'loan_amount': np.random.randint(0,100000,1000),
    'loan_term' : np.random.randint(0,100000,1000),
    'cibil_score': np.random.randint(0,100000,1000),
    'residential_assets_value': np.random.randint(0,100000,1000),
    'commercial_assets_value' : np.random.randint(0,100000,1000),
    'luxury_assets_value' : np.random.randint(0,100000,1000),
    'bank_asset_value' : np.random.randint(0,100000,1000),
}

# Menyimpan data dummy ke dalam dataframe
data_inf = pd.DataFrame(data_inf)
```

```python
# Menambahkan kolom baru 'predict_loan_status' yang berisi model predict
data_inf['predict_loan_status'] = model.predict(data_inf[feature])

# Menampilkan data
data_inf
```

```python
# Menampilkan data yang memprediksi loan status yang approved
data_inf.query('predict_loan_status == 1')
```

```python
# Menampilkan data yang memprediksi loan status yang rejected
data_inf.query('predict_loan_status == 0')
```
